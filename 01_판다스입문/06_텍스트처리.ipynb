```python
# 문자열 매서드 
import pandas as pd
import numpy as np 

ser = pd.Series([
    "Apple_사과", "Banana_바나나", "Cherry_체리", np.nan
], index=["First ", " Second", " Third", "Fourth"])

ser

ser2 = pd.Series([
    "Apple_사과", "Banana_바나나", "Cherry_체리", np.nan
], index=["First ", " Second", " Third", "Fourth"], dtype="string")

ser2
```

```python
ser3 = pd.Series([
    "a1", "b2", "c3", "d4a5", "e6e7", "f8"
], dtype="string")
ser3
```

```python
ser3.str.extractall(r"(\d)").groupby(level=0).agg("".join)
```

```python
ser3.str.extractall(r"([a-z])").groupby(level=0).agg("".join)
```

```python
# 숫자랑 필드랑 컬럼으로 분리 
ser3.str.extract(r"(?P<Letter>[a-z]+)(?P<Digit>\d+)")
```
