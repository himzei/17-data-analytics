```python
# 행인덱스변경 : DataFrame객체.rename(index={기존인덱스: 새인덱스})
# 열이름변경 : DataFrame객체.rename(columns={기존이름: 새이름})

import pandas as pd

matrix = [
    [15, "남", "덕영중"], 
    [17, "여", "수리중"]
]

df = pd.DataFrame(matrix, index=["준서", "예은"], columns=["나이", "성별", "학교"])
df
```
